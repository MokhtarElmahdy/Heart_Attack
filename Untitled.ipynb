```python
def gender_to_catogery( sex ):
    if sex == 1:
        return "male"
    return "female"
```

```python
df1['the gender'] = df1 ['sex']
```

```python
for i in range (df1.shape[0]):
    df1['the gender' ].iloc[i] = gender_to_catogery(df1['sex'].iloc[i])
df1.head()
```

```python
df1['the gender'].value_counts()
```

```python
df1.drop('the gender', axis= 1, inplace=True)
```

```python
x=df1.drop('target',1)
```

```python
y=df1['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```
